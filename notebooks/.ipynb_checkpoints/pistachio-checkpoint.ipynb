```python
train_data = pd.read_csv('../data/train.csv')
train_data
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
for i in train_data.iloc[:,:16]:
    train_data[[i]]=scaler.fit_transform(train_data[[i]])
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

X=train_data.drop('Class',axis=1)
y=train_data.Class
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, shuffle=True, random_state=43)

model = KNeighborsClassifier(n_neighbors=5, p=2, weights='uniform')
model.fit(X_train,y_train)
y_predict=model.predict(X_test)
```

```python
# evaluate your model
from sklearn.metrics import f1_score

prediction = f1_score(y_test,y_predict, average='weighted')
prediction
```
